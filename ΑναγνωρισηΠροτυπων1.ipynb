```python
#adjust file path accordingly
data = np.loadtxt(r'C:\Users\krist\Downloads\ece\8ο\Αναγνωριση προτυπων\Ergastiriakes Askisis\PatRec\DATA\data.dat' )
```

```python
#10x2 -> 5x4, 2 rows for each category
data = np.concatenate([data[i*5:(i+1)*5] for i in range(0,4)], axis=1)
```

```python
fig, axs = plt.subplots(4, figsize=(7,28))

for i,j in enumerate([0,2,4,6]): #for each pair of categories create a scatter subplot

    axs[i].scatter(data[:,j],data[:,j+1], label=f'category {j//2 + 1}',marker = 'D', c='red')    
    axs[i].set_xlabel('feature x')
    axs[i].set_ylabel('feature y')

    axs[i].legend()
```

```python
plt.figure(figsize=(10,10))

for i,j in enumerate([0,2,4,6]): #for each pair of categories create a scatter subplot
           
    plt.scatter(data[:,j],data[:,j+1], label=f'category {j//2 + 1}')    

plt.xlabel('feature x')
plt.ylabel('feature y')
plt.legend()
```

```python
from itertools import combinations

fig, axs = plt.subplots(6, figsize=(7,42))

for i,j in enumerate(combinations([0,2,4,6], 2)): #for each pair of categories create a scatter subplot

    axs[i].scatter(data[:,j[0]],data[:,j[0]+1], label=f'category {j[0]//2 + 1}',marker = 'D', c='red')
    axs[i].scatter(data[:,j[1]],data[:,j[1]+1], label=f'category {j[1]//2 + 1}', c='green')
    
    axs[i].set_xlabel('feature x')
    axs[i].set_ylabel('feature y')

    axs[i].legend()
```
